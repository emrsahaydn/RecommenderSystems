```python
movie_grouped=ratings.groupby('title').agg({'rating':[np.size,np.sum,np.mean]})  #Title ları kendi içerisinde gruplar
```

```python
movie_grouped
```

```python
popular_movies=movie_grouped.sort_values(('rating','mean'),ascending=False)
```

```python
popular_movies
```

```python
grouped_sum=movie_grouped['rating']['sum'].sum()  #rating sütunun içerisindeki oyları toplar.(sum ratinglerin toplamıdır)
```

```python
popular_movies['percentage']=movie_grouped['rating']['sum'].div(grouped_sum)*100 #Her bir filmin yüzdesini bulur.
```

```python
popular_movies.head()
```

```python
popular_movies=popular_movies.sort_values(('percentage'),ascending=False)    #percentage sütununu büyükten küçüğe doğru sıralar
```

```python
popular_movies
```

```python
popular_movies['Rank']=popular_movies['percentage'].rank(ascending=False) #Rank sıralama yapar.
```
