```python
movie_ratings=ratings.pivot_table(index=['user_id'],columns=['title'],values=['rating']) # Matrix factorization
```

```python
movie_ratings #Kullanıcılarla filmler aarsındaki ilişki tablosu
```

```python
print(list(movie_ratings.keys()))
```

```python
movie_ratings.rating['Star Wars (1977)']
```

```python
starwars_ratings=movie_ratings.rating['Star Wars (1977)']
```

```python
starwars_ratings
```

```python
starwars_ratings.value_counts()
```

```python
movie_ratings.rating[['101 Dalmatians (1996)','Star Wars (1977)']].corr(numeric_only=True)
```

```python
movie_ratings.corrwith(starwars_ratings)
```

```python
similar_movies=movie_ratings.corrwith(starwars_ratings)
```

```python
similar_movies=similar_movies.dropna()
```

```python
similar_movies.sort_values(ascending=False)
```

```python
movie_stats=ratings.groupby('title').agg({'rating':[np.size,np.mean]})
```

```python
movie_stats.head()
```

```python
popular_movies=movie_stats[movie_stats['rating']['size']>100]
```

```python
popular_movies.sort_values(('rating','mean'),ascending=False)
```

```python
similar_movies_df=pd.DataFrame(similar_movies,columns=['similarity'])
```

```python
popular_movies.columns=popular_movies.columns.get_level_values(0) #Benzer filmleri getiriyoruz
df=popular_movies.join(similar_movies_df)
```

```python
df.head()
```

```python
df.sort_values(['similarity'],ascending=False)
```
